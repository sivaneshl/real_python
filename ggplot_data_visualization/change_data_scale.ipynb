```python
from plotnine.data import economics

economics
```

```python
from plotnine import ggplot, aes, scale_x_timedelta, labs, geom_line

(
    ggplot(economics)
    + aes(x="date", y="pop")
    + scale_x_timedelta(name="Years since 1970")
    # Using scale_x_timedelta() transforms each pointâ€™s x-value by computing its difference from the oldest date in the dataset
    + labs(title="Population Evolution", y="Population")
    # Uses labs() to set a more descriptive label to the y-axis and the title
    + geom_line()
)
```
