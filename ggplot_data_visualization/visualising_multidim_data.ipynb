```python
from plotnine import ggplot, aes, geom_point, labs

(
    ggplot(mpg) 
    + aes(x='cyl', y='hwy', color='class')
    + labs(
        x='Engine Cylinders',
        y='Miles per Gallon',
        color='Vehicle Class',
        title='Miles per Gallon for Engine Cylinders and Vehicle Class'
    )
    + geom_point()
)
```
